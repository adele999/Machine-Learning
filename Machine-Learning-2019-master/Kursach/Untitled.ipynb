```python
df = pd.read_csv('datadump_S5-PS4.csv')
```

```python
df
```

```python
df['primarysight'] = df['primarysight'].apply(lambda x: '_range' if x=='Acog' else '_close')
df['operator'] = df['operator'] + df['primarysight']

df['objectivelocation'] = df['objectivelocation'].apply(lambda x: '_'+x)
df['mapname'] = df['mapname'] + df['objectivelocation']

df.drop(['primarysight', 'objectivelocation'], axis=1, inplace=True)
```

```python
GAMEMODE = 'PvP – BOMB'
```

```python
df = df[df.gamemode == GAMEMODE]

df.drop('gamemode', axis=1, inplace=True)
```

```python
df = df.merge(df.groupby(['matchid', 'roundnumber']).count()['winrole'].rename('players_in_round').to_frame(),
              how='left',
              left_on=['matchid', 'roundnumber'],
              right_index=True)

df = df[df.players_in_round == 10]
df.drop('players_in_round', axis=1, inplace=True)
```

```python
df = df.reset_index(drop=True)
```

```python
df = pd.concat([df, pd.get_dummies(df.operator)], axis=1)
df.drop('operator', axis=1, inplace=True)
```

```python
df.drop('skillrank', axis=1, inplace=True)
```

```python
df.drop('role', axis=1, inplace=True)
```

```python
df
```

```python
df = df.groupby(['matchid', 'roundnumber', 'mapname', 'winrole', 'endroundreason']).sum().reset_index()
```

```python
df
```

```python
for column in df.select_dtypes(include = ['object']).columns.tolist():
    df[column] = df[column].astype('category')
```

```python
for column in df.select_dtypes(include = ['uint8']).columns.tolist():
    df[column] = df[column].astype('int64')
```

```python
lparams = {}
```

```python
random_search = sklearn.model_selection.GridSearchCV(lightgbm.LGBMClassifier(), lparams, verbose=1, scoring='roc_auc', cv=3)

random_search.fit(df.loc[:,~df.columns.isin(['matchid', 'roundnumber', 'winrole', 'endroundreason'])], df.winrole)
```
